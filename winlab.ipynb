## Load the dataframe

```python
import pandas as pd
```

```python
df=pd.read_csv('wines.csv')
df=df.loc[pd.notna(df['name'])]
df
```

```python
df.columns
```

## How many wines have been given a rating of 100 points?

```python
len(df.loc[df['points']==100])
```

## What is the name of the most expensive wine?

```python
df.loc[df['price']==df['price'].max()]['name'].squeeze()
```

## Calculate a new column where you show the ratings in scale from 0 to 5. Float are allowed.

```python
df['ratings']=df['points']/20
df
```

## Create a price histogram for wines that cost less than 100

```python
df.loc[df['price']<100]['price'].hist()
```

## Plot price horizontally againsts points vertically. Do you think ratings increase by price?

```python
df.plot(kind='scatter',x='price',y='points',figsize=(13,8))
```
